### Defining columns to keep (amount of different value of owner-occupied units)

According to the data dictionary, our dataframe includes many, many columns. While they may be useful for future analysis, let's only keep the columns that we will be using for this session, mainly a selected characteristic of vVALUE, Estimate owner-occupied units
- FIPS:         Identifier                                                                                                     
- DP04_0080E:	Estimate VALUE of Owner-occupied units
- DP04_0081E:	Estimate VALUE of Owner-occupied units Less than $50,000
- DP04_0082E:	Estimate VALUE of Owner-occupied units $50,000 to $99,999
- DP04_0083E:	Estimate VALUE of Owner-occupied units $100,000 to $149,999
- DP04_0084E:	Estimate VALUE of Owner-occupied units $150,000 to $199,999
- DP04_0085E:	Estimate VALUE of Owner-occupied units $200,000 to $299,999
- DP04_0086E:	Estimate VALUE of Owner-occupied units $300,000 to $499,999
- DP04_0087E:	Estimate VALUE of Owner-occupied units $500,000 to $999,999
- DP04_0088E:	Estimate VALUE of Owner-occupied units $1,000,000 or more
- DP04_0089E:	Estimate VALUE of Owner-occupied units Median (dollars)


```python
# define columns to keep; you will have to look at your source's metadata to determine which columns to keep and what they mean
columns_to_keep = ['FIPS',
                   'DP04_0080E',
                   'DP04_0081E',
                   'DP04_0082E',
                   'DP04_0083E',
                   'DP04_0084E',
                   'DP04_0085E',
                   'DP04_0086E',
                   'DP04_0087E',
                   'DP04_0088E',
                   'DP04_0089E']
# add it to a new dataframe
df2 = df[columns_to_keep]
```

```python
# returns list of columns 
columns = list(df2) # this is the same as df.columns.to_list()
columns
```
