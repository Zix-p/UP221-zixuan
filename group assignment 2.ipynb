```python
# read in a geojson file downloaded from the LA Times
caTracts=gpd.read_file('data/tl_2023_06_tract.zip')
caTracts.head()
```

```python
la_tracts = caTracts[caTracts['COUNTYFP'] == '037']
la_tracts.head()
```

```python
la_tracts.plot(figsize=(12,10))
```

```python
la_tracts = la_tracts[['TRACTCE','geometry']]
la_tracts.head()
```

```python
la_tracts['FIPS'] = '06' + '037' + la_tracts['TRACTCE']
```

```python
la_tracts.head()
```

```python
la_tracts['FIPS'] = la_tracts['FIPS'].astype(int)
pd2['FIPS'] = pd2['FIPS'].astype(int)

# Perform the merge
tracts_income = la_tracts.merge(pd2, on="FIPS")
```

```python
# create a new dataframe based on the join
tracts_income=la_tracts.merge(pd2,on="FIPS")
```

```python
tracts_income.head()
```

```python
tracts_income.plot(figsize=(12,10),
                 column='Estimate Households Total',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_income.plot(figsize=(12,10),
                 column='Estimate Households Total',
                 legend=True, 
                 scheme='Quantiles',
                 legend_kwds={
                     'loc':'upper left', 
                     'title':'Estimate Households Total', 
                     'bbox_to_anchor':(1,0.5)},
)
```
